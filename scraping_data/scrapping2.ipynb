```python
from itertools import islice
```

```python
# Fungsi untuk membagi food_names menjadi beberapa bagian dengan ukuran tertentu
def chunked_iterable(iterable, size):
    it = iter(iterable)
    return iter(lambda: list(islice(it, size)), [])

# Membagi food_names menjadi bagian-bagian dengan masing-masing 25 item
chunked_food_names = chunked_iterable(food_names, 25)
```

```python
# List untuk menyimpan informasi makanan dari semua bagian
all_food_information = []

# Memproses setiap bagian
for chunk in chunked_food_names:
    food_information = []
    for food_name in chunk:
        try:
            # Lakukan pencarian
            result = fs.foods_search(food_name)

            # Ambil hanya entri pertama jika ada hasil
            if result:
                food_information.append(result[0])  # Ambil entri pertama dari hasil

            # Delay untuk menghindari rate limit
            time.sleep(5)

        except json.JSONDecodeError as e:
            print(f"Error decoding JSON for {food_name}: {e}")
        except Exception as e:
            print(f"An unexpected error occurred for {food_name}: {e}")

    # Tambahkan hasil dari bagian ini ke all_food_information
    all_food_information.extend(food_information)

# Cetak hasil satu per satu di baris baru
for info in all_food_information:
    print(json.dumps(info, indent=2))
```

```python
all_food_id = []
for food_data in all_food_information:
    try:
        food_id = food_data['food_id']
        all_food_id.append(food_id)
    except (KeyError, TypeError):
        print("food_id not found or invalid data format in all_food_information")
```

```python
import csv

# Nama file CSV
csv_file = "all_food_data_with_servings.csv"

# Kolom utama
main_columns = ['food_name', 'food_id', 'food_type', 'food_url']

# Kolom tambahan dari servings
serving_columns = ['calcium', 'calories', 'carbohydrate', 'cholesterol', 'fat', 'fiber', 'iron', 
                   'potassium', 'protein', 'saturated_fat', 'monounsaturated_fat', 'polyunsaturated_fat', 'serving_description', 'sodium', 'sugar', 'vitamin_a', 'vitamin_c']

# Gabungkan semua kolom
columns = main_columns + serving_columns

# Membuka file CSV untuk menulis data
with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=columns)
    writer.writeheader()

    for food in all_food_data:
        base_row = {col: food.get(col, '') for col in main_columns}
        servings = food.get('servings', {}).get('serving', [])

        # Ubah ke list jika hanya satu serving dalam bentuk dict
        if isinstance(servings, dict):
            servings = [servings]

        for serving in servings:
            row = base_row.copy()
            row.update({col: serving.get(col, '') for col in serving_columns})
            writer.writerow(row)

print(f"{csv_file} berhasil dibuat dengan semua food dan servings-nya 🎉")
```

```python
# Membaca file CSV yang telah dibuat
df = pd.read_csv(csv_file)
excel_file = "all_food_data_with_servings.xlsx"
df.to_excel(excel_file, index=False)

print(f"{excel_file} berhasil dibuat dalam format Excel ✅")
```
